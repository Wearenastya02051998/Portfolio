–ü–æ–ª—É—á–∏–ª–æ—Å—å –æ—á–µ–Ω—å –º–Ω–æ–≥–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º, –æ—Å—Ç–∞–≤–∏–º —Ç–æ–ø 10 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º –∏ –ø–æ—Å–º—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –ø–æ –≥–æ–¥–∞–º —Ç–æ–ª—å–∫–æ –¥–ª—è –Ω–∏—Ö. 

````python
total_sum = games.groupby(['year_of_release', 'platform'])['total'].sum().reset_index()

# –°–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø–æ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º –∏ –ø–æ—Å—á–∏—Ç–∞—Ç—å –æ–±—â—É—é —Å—É–º–º—É –¥–ª—è –∫–∞–∂–¥–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã
platform_total_sum = total_sum.groupby('platform')['total'].sum().reset_index()

# –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –ø–æ —É–±—ã–≤–∞–Ω–∏—é –æ–±—â–µ–π —Å—É–º–º—ã –∏ –≤—ã–±—Ä–∞—Ç—å —Ç–æ–ø 10
top_platforms = platform_total_sum.nlargest(10, 'total')['platform']

# –§–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å –∏—Å—Ö–æ–¥–Ω—ã–π DataFrame, —á—Ç–æ–±—ã –æ—Å—Ç–∞–≤–∏—Ç—å —Ç–æ–ª—å–∫–æ —Å—Ç—Ä–æ–∫–∏, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ —Ç–æ–ø 10 –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º
top_games = games[games['platform'].isin(top_platforms)]

# –ü–æ—Å—á–∏—Ç–∞—Ç—å —Å—É–º–º—É –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ 'total' –¥–ª—è –∫–∞–∂–¥–æ–π —É–Ω–∏–∫–∞–ª—å–Ω–æ–π –∫–æ–º–±–∏–Ω–∞—Ü–∏–∏ –≥–æ–¥–∞ –≤—ã–ø—É—Å–∫–∞ –∏ —Ç–æ–ø 10 –ø–ª–∞—Ç—Ñ–æ—Ä–º
top_total_sum = top_games.groupby(['year_of_release', 'platform'])['total'].sum().reset_index()

# –°–æ–∑–¥–∞—Ç—å —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º plotly
fig = px.bar(top_total_sum, x='year_of_release', y='total', color='platform', text='platform',
             title='–°—É–º–º–∞ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ "total" –¥–ª—è —Ç–æ–ø 10 –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –≥–æ–¥–∞–º –≤—ã–ø—É—Å–∫–∞',
             labels={'year_of_release': '–ì–æ–¥ –≤—ã–ø—É—Å–∫–∞', 'total': '–°—É–º–º–∞ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–µ "total"'})

# –ù–∞—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–¥—Å–∫–∞–∑–∫–∏
fig.update_traces(hovertemplate='–ì–æ–¥ –≤—ã–ø—É—Å–∫–∞: %{x}<br>–ü–ª–∞—Ç—Ñ–æ—Ä–º–∞: %{text}<br>–°—É–º–º–∞: %{y}')

# –ü–æ–∫–∞–∑–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫
fig.show()
````

### –ü—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑

### –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ;

–ë—É–¥–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω —Ç–µ—Å—Ç –°—Ç—é–¥–µ–Ω—Ç–∞.
–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ - —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC —Ä–∞–≤–Ω—ã.
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ - —Å—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –Ω–µ —Ä–∞–≤–Ω—ã.
–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç alpha –±—É–¥–µ—Ç —Ä–∞–≤–µ–Ω 0.01.

````python
user_xone = filtered_games.query('platform == "XOne"')
user_xone = user_xone[user_xone['user_score'].notna()]
user_xone['user_score'].describe()
````

````python
user_pc = filtered_games.query('platform == "PC"')
user_pc = user_pc[user_pc['user_score'].notna()]
user_pc['user_score'].describe()
````

````python
alpha = 0.01

results = st.ttest_ind(user_xone['user_score'], user_pc['user_score'])

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if (results.pvalue < alpha):
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
````
